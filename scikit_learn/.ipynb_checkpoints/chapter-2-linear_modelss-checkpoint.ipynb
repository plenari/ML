## Using ridge regression to overcome linear regression's shortfalls

```python
from sklearn.datasets import make_regression
import numpy as np
reg_data, reg_target = make_regression(n_samples=2000,n_features=3, effective_rank=2, noise=10)
reg_data.shape,reg_target.shape
```

```python
reg_data.ndim#秩 rank 
```

```python
from sklearn .linear_model import Ridge
r=Ridge()
n=1000
lendata=len(reg_data)
size=np.int(0.75*lendata)
subsample=lambda:np.random.choice(np.range(0,lendata),size=size)
coefs_r=np.ones((n,3))
#np.mean(coefs-coefs_r,axis=0)
```

```python
r.fit(x,y)
```

```python
plot(r.predict(x),y,'*')
```

### 优化岭回归的参数
RidgeCV(alphas=(0.1, 1.0, 10.0), fit_intercept=True, normalize=False, scoring=None, cv=None, gcv_mode=None, store_cv_values=False

```python
from sklearn .linear_model import RidgeCV
rcv=RidgeCV(alphas=np.array([.1,.5,.8,1.0]))
rcv.fit(x,y)
```

```python
rcv.alpha_#this is the best alpha
```

```python
alphas_test = np.linspace(-0.01, 1)
rcv3=RidgeCV(alphas=alphas_test,store_cv_values=True)
rcv3.fit(x,y)
```

```python
plot(alphas_test, rcv3.cv_values_.mean(axis=0))
```

```python
plot(alphas_test, rcv3.cv_values_.var(axis=0))
```

### 岭回归可以自定义损失函数


```python
def MAD(target,predictions):
    absolute_deviation = np.abs(target - predictions)
    return absolute_deviation.mean()
from sklearn import metrics#
MAD=metrics.make_scorer(MAD,greater_is_better=False)
rcv4=RidgeCV(alphas=alphas_test,store_cv_values=True,scoring=MAD)
rcv4.fit(reg_data,reg_target)
```

```python
plot(alphas_test, rcv4.cv_values_.mean(axis=0))
```

## Taking a more fundamental approach to regularization with LARS
Least-angle regression (LARS) is a regression technique that is well suited for high-dimensional
problems, that is, p >> n, where p denotes the columns or features and n is the number
of samples.

```python
x,y=make_regression(200,500,n_informative=10,noise=2)
x.shape,y.shape
```

```python
from sklearn.linear_model import Lars
lars=Lars(n_nonzero_coefs=10)
lars.fit(x,y)
np.sum(lars.coef_!=0)
```

```python
from sklearn.linear_model import LarsCV
larcv=LarsCV()
larcv.fit(x,y)
np.sum(larcv.coef_!=0)
```

## Using linear methods for classification-logistic regression
Linear models can actually be used for classification tasks. This involves fitting a linear model
to the probability of a certain class, and then using a function to create a threshold at which
we specify the outcome of one of the classes.

## Directly applying Bayesian ridge regression
Ridge and lasso regression can both be understood through a Bayesian lens as opposed to
an optimization lens. Only Bayesian ridge regression is implemented by scikit-learn, but in the
How it works... section, we'll look at both cases.

```python
from sklearn.datasets import make_regression
x, y = make_regression(1000, 10, n_informative=2, noise=20)
```

```python
from sklearn.linear_model import BayesianRidge
br=BayesianRidge()
```

```python
br.fit(x,y)
br.coef_
```

```python
br_alphas=BayesianRidge(alpha_1=10,lambda_1=10)#不是很懂
br_alphas.fit(x,y)
br_alphas.coef_
```

## 较弱的集成学习？

Gradient boosting regression is a technique that learns from its mistakes. Essentially, it tries to fit a bunch of weak learners. There are two things to note:
1,Individually, each learner has poor accuracy, but together they can have very good accuracy
2, They're applied sequentially, which means that each learner becomes an
expert in the mistakes of the prior learner

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR
gbr=GBR()
gbr.fit(x,y)
```

```python
gbr_pred=gbr.predict(x)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
lr_pred=lr.predict(x)
```

```python
gbr_residuals=y-gbr_pred
lr_residuals=y-lr_pred
plot(gbr_residuals,'ro')
plot(lr_residuals,'b*')
```

```python
#np.percentile(a, q, axis=None, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
#计算百分位
np.percentile(gbr_residuals,[2.5,97.5])
```

```python
np.percentile(lr_residuals,[2.5,97.5])
```
