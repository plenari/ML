### Probabilistic clustering with Gaussian Mixture Models
混合高斯模型
以男女生身高为例
#### np.random.normal(loc=0.0, scale=1.0, size=None),第一个是u,第二个 是方差

累积分布函数，又叫分布函数，是概率密度函数的积分，能完整描述一个实随机变量X的概率分布。一般以大写“CDF”（Cumulative Distribution Function）标记。
当概率密度函数存在的时候，累积分布函数是概率密度函数的积分。概率密度函数一般以大写“PDF”（Probability Density Function）标记[1]。


```python
import numpy as np
N = 1000
in_m = 72
in_w = 66
s_m = 2
s_w = s_m
m = np.random.normal(in_m, s_m, N)#create normal data
w = np.random.normal(in_w, s_w, N)
from matplotlib import pyplot as plt
f, ax = plt.subplots(figsize=(7, 5))
ax.set_title("Histogram of Heights")
ax.hist(m, alpha=.5, label="Men");
ax.hist(w, alpha=.5,label='Women')
ax.legend()
```

```python
random_sample = np.random.choice([True, False], size=m.size)#从真假里面随机选择。
m_test = m[random_sample]
m_train = m[~random_sample]
w_test = w[random_sample]
w_train = w[~random_sample]
```

```python
from scipy import stats
m_pdf = stats.norm(m_train.mean(), m_train.std())#用训练数据构造正态分布
w_pdf = stats.norm(w_train.mean(), w_train.std())
```

```python
m_pdf.pdf(m[0])
```

```python
w_pdf.pdf(m[0])
```

```python
guesses_m = np.ones_like(m_test)
guesses_m[m_pdf.pdf(m_test) < w_pdf.pdf(m_test)] = 0
```

```python
plot(guesses_m)
```

```python
guesses_w = np.ones_like(w_test)
guesses_w[m_pdf.pdf(w_test) > w_pdf.pdf(w_test)] = 0
guesses_w.mean()
plot(guesses_w)
```

### Using KMeans for outlier detection


```python
X,labels=make_blobs(100,centers=1)
kmeans=KMeans(n_clusters=1)
kmeans.fit(X)
```

```python
plot(X[:,0],X[:,1],'b.')
plot(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],'r*')
```

```python
distances = kmeans.transform(X)
```

```python
sorted_idx = np.argsort(distances.ravel())[::-1][:5]
sorted_idx
```

```python
new_X = np.delete(X, sorted_idx, axis=0)
#删除那些误差较大的点
```

## Using k-NN for regression
K-Nearest Neighbors


```python
from sklearn import datasets
iris=datasets.load_iris()
iris.feature_names
X=iris.data
y=iris.target
```

```python
from sklearn.linear_model import LinearRegression as LR
lr=LR()
lr.fit(X,y)
print('MSE',np.power(y-lr.predict(X),2).mean())
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNR
knnr=KNR(n_neighbors=10)
knnr.fit(X,y)
print('MSE',np.power(y-lr.predict(X),2).mean())
```
