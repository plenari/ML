# K-nn近邻

### 欧式距离（p=2）
    两点之间的距离
### 曼哈顿距离（p=1）
    两点不同维度距离的绝对值之和
### p无穷大
    两点不同维度距离的绝对值最大

## K的选择
    K值较小，意味着只有最近的几个点进行预测。会增大误差，使模型非常复杂
    K值较大，这样分到同一个类别的相似性会减小
    通常用交叉验证的方法选择一个合适的k

```python
import numpy as np
from sklearn import neighbors, datasets
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

#### neighbors.KNeighborsClassifier?
    neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=1, **kwargs)
    Docstring:     
    Classifier implementing the k-nearest neighbors vote.

```python
iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target
```

```python
#cmap1=plt.get_cmap()
cmap_l = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_b = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
clf=neighbors.KNeighborsClassifier(n_neighbors=5)
clf.fit(X,y)
```

```python
y_fit=clf.predict(X)
print('correct {}'.format(np.mean(y_fit==y)))
```

### 实际

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm)
plt.show()
```

### 预测

```python
plt.scatter(X[:,0],X[:,1],c=y_fit,cmap=cmap_b)
plt.show()
```

### 不符

```python
index_not_fit=np.where(y_fit!=y)
```

```python
x_error=X[index_not_fit]
```

### 画mesh图片

```python
xmin,ymin=X[:,0].min()-1,X[:,1].min()-1
xmax,ymax=X[:,0].max()+1,X[:,1].max()+1
length=300
```

```python
XX,YY=np.meshgrid(np.linspace(xmin,xmax,length),np.linspace(ymin,ymax,length))
new_y=clf.predict(np.c_[XX.flatten(),YY.flatten()]).reshape(XX.shape)
```

```python
plt.figure()
plt.pcolormesh(XX,YY,new_y,cmap=cmap_l)
plt.scatter(X[:,0],X[:,1],c=y_fit,cmap=cmap_b)
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
plt.title('3')
plt.show()
```
